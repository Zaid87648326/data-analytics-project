# presentation du dataset:
Ce jeu de données semble être extrait d'un classement du bonheur mondial. Chaque ligne représente un pays avec des informations telles que le rang de bonheur, le score de bonheur, le PIB par habitant, la santé (espérance de vie), la liberté, la confiance dans le gouvernement, la générosité, etc. Ces données sont souvent utilisées dans le domaine de la recherche en sciences sociales pour étudier les facteurs qui contribuent au bien-être des individus à l'échelle mondiale. En français, on pourrait les appeler "Les données du Bonheur Mondial".

# objectif de projet:
L'objectif de ce projet est de classifier les pays en fonction de leur score de bonheur, de leur generosite et de la confiance qu ils accordent a leur gouvernement .
en utilisant l algorithme de k-means clustering, nous cherchons à identifier des groupes de pays ayant des caractéristiques similaires dans ces trois domaines. Cette classification peut aider à mieux comprendre les facteurs qui influencent le bien-être des pays et à identifier des modèles ou des tendances qui pourraient être utilisés pour informer les politiques publiques

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv("2015.csv")
data.head(5)
```

```python
from sklearn import cluster
from sklearn.cluster import KMeans

# choix es colonnes pertinentes
X = data[['Happiness Score', 'Generosity', 'Trust (Government Corruption)']]
y = data['Region']
# efectuation du K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisation
plt.figure(figsize=(10, 7))
sns.scatterplot(data=data, x='Happiness Score', y='Generosity', hue='Cluster', palette='viridis')
plt.title('Clusters of Countries based on Happiness Score and Generosity')
plt.show()
```

**INTERPRETATION :**

la représentation est une visualisation de la répartition de la générosité (« generosity ») en fonction du score de bonheur (« Happiness score ») est la suivante :

La visualisation montre une tendance claire selon laquelle les pays ayant un score de bonheur élevé (« 2 ») ont tendance à être plus généreux en moyenne à travers le monde. Ces pays sont généralement représentés par des scores plus élevés sur l’axe vertical, ce qui correspond à une valeur de générosité plus élevée. En revanche, les pays ayant de faibles niveaux de bonheur (« 0 » et « 1 ») ont tendance à être moins généreux, comme l’indiquent les points regroupés au bas de l’axe vertical, avec des valeurs de générosité plus faibles.

 Il peut y avoir des exceptions à cette tendance générale, comme les pays qui ont un score de bonheur relativement faible mais qui font preuve d’une plus grande générosité, ou vice versa.
